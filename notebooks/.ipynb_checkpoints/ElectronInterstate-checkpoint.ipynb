```julia
# Import modules and subroutines:

include("../subroutines/Subroutines.jl");
```

```julia
# Load the chemical data into julia as an array of chemical data structs:
# (one struct obtained for each molecular geometry as set up in the config file)

cdata_list = ReadIn("../datasets/pyscf_data/h2o_2p2A_sto-3g_021423%124204.hdf5");
chemical_data = cdata_list[1]
PrintChemData(chemical_data)
```

```julia
# Run a DMRG calculation to get the entanglement profile of the ground state:

dmrg1 = GenSubspace(chemical_data, 1, psi_maxdim=20, sweep_num=40, sweep_noise=(1e-3,1e-3,1e-3,1e-4,1e-4,1e-4,1e-5,1e-5,1e-5,1e-6,1e-6,1e-6,1e-7));
dmrg1.ord_list = [collect(1:dmrg1.chem_data.N_spt)]
GenStates!(dmrg1)
GenPermOps!(dmrg1)
GenSubspaceMats!(dmrg1)

println(dmrg1.H_mat[1,1]+dmrg1.chem_data.e_nuc)

"""
for k=1:3
    _,_, Ipq0 = MutualInformation(dmrg1.psi_list[1], dmrg1.chem_data)
    dmrg1.ord_list = InfDistAnnealing(
        Ipq0, 
        1, 
        gp,
        verbose=true
    )

    GenStates!(dmrg1)
    GenPermOps!(dmrg1)
    GenSubspaceMats!(dmrg1)
    
    println(dmrg1.H_mat[1,1]+dmrg1.chem_data.e_nuc)
end
"""

#biparts, entrops = ComputeBipartites(dmrg1, state=1);
```

```julia
S1, S2, Ipq = MutualInformation(dmrg1.psi_list[1], dmrg1.chem_data)

display(Ipq)

OrbitalGraphPlot(Ipq, multiplier=3e1)
```

```julia
gp = GeomParameters(
    costfunc = "simple",
    a_maxiter=300000,
    eta=-2,
    shrp=2.0,
    a_alpha=1e1
)
```

```julia
# Get quasi-optimal orderings via simulated annealing:

tnqe0 = GenSubspace(chemical_data, 5, psi_maxdim=3, thresh="projection", eps=1e-12);

tnqe0.ord_list = InfDistAnnealing(
    Ipq, 
    tnqe0.mparams.M, 
    gp,
    verbose=true
)
```

```julia
graph_mat_list = []

for (j,ord) in enumerate(tnqe0.ord_list)
    push!(graph_mat_list, GraphMat(ord, directed=false) )
end

for j=1:length(graph_mat_list)
    println("EXCITATION LEVEL: $(j-1)")
    OrbitalGraphPlot(graph_mat_list[j], multiplier=2.0)
end
```

```julia
opt_sweeps3 = Sweeps(40)
maxdim!(opt_sweeps3,3)
mindim!(opt_sweeps3,3)
cutoff!(opt_sweeps3,1e-12)
setnoise!(opt_sweeps3, 1e-3,1e-3,1e-3,1e-4,1e-4,1e-4,1e-5,1e-5,1e-5,1e-6,1e-6,1e-6,1e-7)

GenHams!(tnqe0)

GenStates!(tnqe0, randomize=true, verbose=true)

_, tnqe0.psi_list[end] = dmrg(tnqe0.ham_list[end], tnqe0.psi_list[end], opt_sweeps3, outputlevel=0);
```

```julia
GenPermOps!(tnqe0, verbose=true)
GenSubspaceMats!(tnqe0)
SolveGenEig!(tnqe0, verbose=true)
```

```julia
tnqe1 = copy(tnqe0)

op1 = OptimParameters(
    maxiter=1, 
    numloop=1, 
    noise=[0.0], 
    sd_thresh="inversion", 
    sd_eps=1.0e-8,
    delta=[0.0],
    afunc="flat",
    ttol=10.0,
    theta=0.0,
    sd_method="triple_geneig",
    sd_penalty=1.0-1e-6
)

TwoSitePairSweep!(
    tnqe1,
    op1,
    jpairs=[[1,2],[2,3],[3,4],[4,5],[1,5]],
    verbose=true
)

TwoSitePairSweep!(
    tnqe1,
    op1,
    jpairs=[[1,2],[2,3],[3,4],[4,5],[1,5]],
    verbose=true
)

OneSitePairSweep!(
    tnqe1,
    op1,
    jpairs=[[1,2],[2,3],[3,4],[4,5],[1,5]],
    verbose=true
)
```
