```julia
# Load the chemical data into julia as an array of chemical data structs:
# (one struct obtained for each molecular geometry as set up in the config file)

cdata_list = ReadIn("../datasets/pyscf_data/h6_0.95A_sto-3g_052323%1850.hdf5");
chemical_data = cdata_list[1]
PrintChemData(chemical_data)
```

```julia
fci_array = FCIArray(chemical_data)
fci_mps = MPS(fci_array, siteinds("Electron", 6), cutoff=1e-16, maxdim=2^16);
```

```julia
S1, S2, Ipq = MutualInformation(fci_mps, collect(1:chemical_data.N_spt), chemical_data)

display(Ipq)

OrbitalGraphPlot(Ipq, multiplier=1.5e1)
```

```julia
gp = GeomParameters(
    maxiter=300000,
    eta=-2,
    shrp=2.0,
    a_alpha=1e0
)

op1 = OptimParameters(
    maxiter=1, 
    numloop=1, 
    noise=[0.0], 
    sd_thresh="inversion", 
    sd_eps=1.0e-8,
    delta=[0.0],
    ttol=10.0,
    theta=0.0,
    alpha=1e2,
    sd_method="triple_geneig",
    sd_penalty=1.0-1e-5
)

op2 = OptimParameters(
    maxiter=1, 
    numloop=1, 
    noise=[0.0], 
    sd_thresh="inversion", 
    sd_eps=1.0e-8,
    delta=[0.0],
    ttol=10.0,
    theta=0.0,
    alpha=1e10,
    gamma=0.5,
    sd_method="triple_geneig",
    sd_penalty=1.0-1e-6,
    sd_swap_penalty=1.0-1e-3
)
```

```julia
# Get quasi-optimal orderings via simulated annealing:

tnqe0 = GenSubspace(chemical_data, 6, psi_maxdim=4, thresh="projection", eps=1e-12);

#tnqe0.ord_list = [randperm(tnqe0.chem_data.N_spt) for i=1:tnqe0.mparams.M]
display(tnqe0.ord_list)

tnqe0.ord_list = InfDistAnnealing(
    Ipq, 
    tnqe0.mparams.M, 
    gp,
    verbose=true
)

display(tnqe0.ord_list)

GenHams!(tnqe0)
GenStates!(tnqe0, randomize=true, verbose=true)
GenPermOps!(tnqe0)
GenSubspaceMats!(tnqe0)
SolveGenEig!(tnqe0)

TwoSitePairSweep!(
    tnqe0,
    op1,
    verbose=true
)
```

```julia
nomps0 = GenSubspace(chemical_data, 6, psi_maxdim=4, thresh="projection", eps=1e-12);

opt_ord = InfDistAnnealing(
    Ipq, 
    1, 
    gp,
    verbose=true
)[1]

nomps0.ord_list = [deepcopy(opt_ord) for i=1:nomps0.mparams.M]

GenHams!(nomps0)
GenStates!(nomps0, randomize=true, verbose=true)
GenPermOps!(nomps0)
GenSubspaceMats!(nomps0)
SolveGenEig!(nomps0)

TwoSitePairSweep!(
    nomps0,
    op1,
    verbose=true
)
```

```julia
psi = deepcopy(tnqe2.psi_list[1]);

T = psi[2] * psi[3];

oht_list = OneHotTensors(T);
```

```julia
t_vec = randn(length(oht_list));

normalize!(t_vec)

T_new = sum([t_vec[k]*oht_list[k] for k=1:length(oht_list)]);

linds = commoninds(T_new, psi[2]);

U,S,V = svd(T_new, linds);

L = size(S.tensor, 1)

s_vec = [S[i,i] for i=1:L]

s2_vec = s_vec .^ 2

vn_entropy = sum(-log.(s2_vec) .* s2_vec)

println(vn_entropy)

println(sum(reverse(sort(s2_vec))[1:4]))

println("##############")

fswap = BuildFermionicSwap(tnqe2.sites, 2; dim=4);

T_new_prime = T_new * fswap
noprime!(T_new_prime);

Up,Sp,Vp = svd(T_new_prime, linds);

Lp = size(Sp.tensor, 1)

s_vec_p = [Sp[i,i] for i=1:Lp]

s2_vec_p = s_vec_p .^ 2

vn_entropy_p = sum(-log.(s2_vec_p) .* s2_vec_p)

println(vn_entropy_p)

println(sum(reverse(sort(s2_vec_p))[1:4]))
```
